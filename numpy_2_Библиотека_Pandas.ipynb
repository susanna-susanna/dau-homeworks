###Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings.head()
```

```python
ratings.info()
```

```python
# оставляем только тех, у которых 5-ки
only_5 = ratings.query('rating == 5')
only_5.info()
```

```python
# ищем через .value_counts()
show_top_1 = only_5['movieId'].value_counts()
show_top_1.head()
```

```python
# ищем через .groupby() 
show_top_2 = only_5.groupby('movieId').count().sort_values(by="rating", ascending=False)
show_top_2.head()
```

```python
show_top_2.describe()
```

###Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power[power['country'].str.contains('esto', case=False)]['country'].unique()
# данный фильтр берёт строки, в которых названия стран точно совпадают с одной из указанных
```

```python
power['country'].str.contains('Estonia|Latvia|Lithuania')
```

```python
# чтоб не по одному имени искать, а сразу всё, что мне нужны (строки,содеращие подстроку),
# то подойдёт такой фильтр:
power[power['country'].str.contains('Estonia|Latvia|Lithuania', case=False)]['country'].unique()
```

```python
# фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# () | (() | () & ())

# вот это отдельный фильтр (data['country']=='Russian Federation')
# это отдельный фильтр (data['country']=='Belarus')

filtered_countries = power[ (power['country'] =='Latvia') | (power['country'] =='Lithuania') | (power['country'] =='Estonia') & (power['category'] == 4)]

filtered_countries.head()
```

```python
pribaltica = power[power['country'].isin(['Latvia', 'Lithuania', 'Estonia'])]
pribaltica.head()
```

```python
baltica_cat = pribaltica[pribaltica['category'].isin([4, 12, 21])]
baltica_cat.head()
```

```python
baltica_cat_years = baltica_cat[(baltica_cat['year'] >= 2005) & (baltica_cat['year'] <= 2010) & (baltica_cat['quantity'] > 0)]
baltica_cat_years.head()
```

```python
# можно немного сократить код, используя для дат метод between
baltica_cat_years_1 = baltica_cat['year'].between(2005, 2010)
baltica_cat_years_1.head()
```

```python
baltica_cat_years.describe()
```

```python
baltica_cat_years['quantity'].sum()
```
